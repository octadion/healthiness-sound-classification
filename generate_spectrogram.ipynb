```python
from IPython.display import Audio
import librosa
from librosa import display
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
import pylab
from PIL import Image
from matplotlib.pyplot import imshow
import os
%matplotlib inline
```

```python
WAV_DIR = 'wav_files/'
IMG_DIR = 'spectrogram_images/'
wav_files = os.listdir(WAV_DIR)
```

```python
for f in os.listdir(WAV_DIR):
    try:
        # Read wav-file
        y, sr = librosa.load(os.path.join(WAV_DIR, f), sr=22050)  # Menggunakan sampling rate default 22,050 Hz

        # Compute spectrogram
        M = librosa.feature.melspectrogram(y=y, sr=sr,
                                           fmax=sr/2,  # Frekuensi maksimum yang digunakan dalam skala MEL
                                           n_fft=2048,
                                           hop_length=512,
                                           n_mels=96,  # Disesuaikan dengan Google Large-scale audio CNN paper
                                           power=2)  # Power = 2 mengacu pada amplitudo kuadrat

        # Power dalam skala dB
        log_power = librosa.power_to_db(M, ref=np.max)  # Konversi ke skala dB (log)

        # Plot spektrogram
        plt.figure(figsize=(5, 5))
        plt.axis('off')
        plt.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[])  # Menghapus pinggiran putih
        librosa.display.specshow(log_power, cmap='jet')
        plt.savefig(os.path.join(IMG_DIR, f[:-4] + '.jpg'), bbox_inches=None, pad_inches=0)
        plt.close()

    except Exception as e:
        print(f, e)
        pass
```
